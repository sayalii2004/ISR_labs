```python
import matplotlib.pyplot as plt
from skimage.io import imread
from skimage.color import rgb2gray
from skimage.filters import threshold_otsu, gaussian
```

```python
# helper function

def show_img(img, title='Image'):
    plt.imshow(img, cmap='gray')
    plt.title(title)
    plt.axis('off')
    plt.show()
```

```python
def show_images(images, titles):
  n=len(images)
  plt.figure(figsize=(20,20))
  for i in range(n):
    plt.subplot(1,n,i+1)
    if images[i].ndim == 3:
      plt.imshow(images[i], cmap='gray')
    else:
      plt.imshow(images[i])
    plt.title(titles[i])
    plt.axis('off')
  plt.show()
```

```python
image = imread("/content/flower.jpg")
show_img(image,"original image")
```

```python
# red and yellow intensity filters

red = image.copy()
yellow = image.copy()

red[:,:,(1,2)]=0
yellow[:,:,2]=0

show_images([image, red, yellow], ["original", "red", "yellow"])
```

```python
# convert to grayscale

gray_image = rgb2gray(image)
show_img(gray_image, "grayscale image")
```

```python
# grayscale image

print("coloured image shape :",image.shape)
print("GrayScale image shape :",gray_image.shape)
```

```python
# otsu threshold

thresh = threshold_otsu(gray_image)
binary = gray_image > thresh
show_img(binary, "Ostu binary image")
```

```python
# Guassian blur

blurred_image = gaussian(gray_image, sigma=3)
show_img(blurred_image, "Blurred image")
```

```python
# plot histogram

plt.hist(gray_image.ravel(), bins=256)
plt.show()
```
